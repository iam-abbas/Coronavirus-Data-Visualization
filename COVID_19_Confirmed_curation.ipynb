```
import pandas as pd
import numpy as np
```

```
data = pd.read_csv('covid_19_data.csv')
data.head()
```

```
data['ObservationDate'] = data['ObservationDate'].apply(pd.to_datetime())
# data['Country'] = data['Country/Region'].replace('Mainland China', 'China')
data['Country'] = data['Country/Region']
data = data.drop(columns=['SNo', 'Province/State', 'Last Update', 'Deaths', 'Country/Region', 'Recovered'])
data.head()
```

```
data = data.groupby(['ObservationDate', 'Country']).sum().reset_index()
data.tail(25)
```

```
dates = [str(x)[:10] for x in data['ObservationDate'].unique()]
countries = [x for x in data['Country'].unique()]
```

```
cases = pd.DataFrame(columns=dates)
cases.insert(0, 'Country', countries)
```

```
cases.head()
```

```
cases = cases.set_index("Country")
cases.head()
```

```
for country in countries:
  for date in dates:
    try:
      case = int(data['Confirmed'].loc[(data['ObservationDate'] == date) & (data['Country'] == country)])
    except:
      case = 0
    
    cases.at[country, date] = case
```

```
cases.head(30)
```

```
cases.to_csv("data_final.csv", encoding='utf-8', index=True)
```

```
cases_without_china = cases.drop("Mainland China")
```
